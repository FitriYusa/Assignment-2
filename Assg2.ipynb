#### Handling missing values

```python
df.isna().sum()
```

```python
df = df.fillna(df.mean(numeric_only=True))
df['Address'] = df['Address'].fillna('Unknown')
display(df.isna().sum())
```

#### Variable Transformation

```python
# Convert numerical columns stored as object to float
# Convert columns with currency and percentage symbols to float
num_string_cols = [
    'Cost Price',
    'Retail Price',
    'Profit Margin',
    'Sub Total',
    'Discount %',
    'Discount $',
    'Order Total',
    'Shipping Cost',
    'Total'
]

for col in num_string_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace('%', '', regex=False)
        .str.replace(',', '', regex=False)
        .astype(float)
    )

# Convert percentage value to decimal    
if '%' in col:
    df[col] = df[col] / 100
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
```

```python
# Check data types after conversion
df.dtypes
```

#### Feature engineering (Categorization/feature selection)

```python
#Clustering Features Engineering

# Profit margin: Markup percentage from cost to retail price
df['Markup_Percentage'] = (df['Retail Price'] - df['Cost Price']) / df['Cost Price']

# Profit per unit 
df['Profit per Unit'] = df['Retail Price'] - df['Cost Price']

# Shipping impact: Cost of shipping relative to the total price
df['Shipping_Impact_Score'] = df['Shipping Cost'] / df['Order Total']

# Order size banding
df['Order Size Band'] = pd.cut(
    df['Order Quantity'],
    bins=[0,5,20,float('inf')],
    labels=['Small','Medium','Large']
    )
```
